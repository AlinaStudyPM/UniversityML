```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
data = pd.read_csv('FLATS_FOR_11_PREZ.csv', encoding='utf-8', delimiter=';', index_col=0)
data
```

```python
X = data[['SQUARE_METERS', 'METRO_DISTANCE']]
y = data['PRICE']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Строим модель линейной регрессии
model_lin = LinearRegression()
model_lin.fit(X_train, y_train)
```

```python
#Предсказанные значения для тестовых данных
y_pred_lin = model_lin.predict(X_test)
```

```python
#Строим модель квадратичной регрессии
quad = PolynomialFeatures(degree=2)
X_quad = quad.fit_transform(X_train)
model_quad = LinearRegression()
model_quad.fit(X_quad, y_train)
```

```python
#Предсказанные значения для тестовых данных
X_test_quad = quad.transform(X_test)
y_pred_quad = model_quad.predict(X_test_quad)
```

```python
#Вычисляем значения метрик для линейной модели
mae_lin = mean_absolute_error(y_test, y_pred_lin)
rmse_lin = np.sqrt(mean_squared_error(y_test, y_pred_lin))
mpe_lin = np.mean((y_test - y_pred_lin) / y_test) * 100
mape_lin = np.mean(np.abs(y_test - y_pred_lin) / y_test) * 100
ad_lin = np.sum(np.abs(y_test - y_pred_lin))
mad_lin = np.mean(np.abs(y_test - y_pred_lin))
r2_lin = r2_score(y_test, y_pred_lin)
```

```python
#Вычисляем значения метрик для квадратичной
mae_quad = mean_absolute_error(y_test, y_pred_quad)
rmse_quad = np.sqrt(mean_squared_error(y_test, y_pred_quad))
mpe_quad = np.mean((y_test - y_pred_quad) / y_test) * 100
mape_quad = np.mean(np.abs(y_test - y_pred_quad) / y_test) * 100
ad_quad = np.sum(np.abs(y_test - y_pred_quad))
mad_quad = np.mean(np.abs(y_test - y_pred_quad))
r2_quad = r2_score(y_test, y_pred_quad)
```

```python
print("Метрики для линейной модели:")
print(f"MAE: {mae_lin:.4f}")
print(f"RMSE: {rmse_lin:.4f}")
print(f"MPE: {mpe_lin:.4f}%")
print(f"MAPE: {mape_lin:.4f}%")
print(f"AD: {ad_lin:.4f}")
print(f"MAD: {mad_lin:.4f}")
print(f"Коэффициент детерминации (R²): {r2_lin:.4f}")
```

```python
print("\nМетрики для полиномиальной модели:")
print(f"MAE: {mae_quad:.4f}")
print(f"RMSE: {rmse_quad:.4f}")
print(f"MPE: {mpe_quad:.4f}%")
print(f"MAPE: {mape_quad:.4f}%")
print(f"AD: {ad_quad:.4f}")
print(f"MAD: {mad_quad:.4f}")
print(f"Коэффициент детерминации (R²): {r2_quad:.4f}")
```
